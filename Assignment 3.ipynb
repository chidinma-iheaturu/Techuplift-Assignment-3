# Word Counter

```python
def count_words():
    '''
    Count the number of words in a given sentence.
    '''
    # Input from the user
    
    sentence = str(input("Please Enter your sentence: "))
    
    # Split the sentence into words
    
    word = sentence.split()
    
    # Count the number of words
    
    word_count = len(word)
    
    print("The Number of words in the sentence is:", word_count)
```

```python
count_words()
```

# Banking System

```python
class Wallet:
    def __init__(self, account_owner, balance=0):
        self.account_owner = account_owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ₦{amount}. New balance: ₦{self.balance}")
        else:
            print("Invalid deposit amount. Please enter a positive value.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ₦{amount}. New balance: ₦{self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def check_balance(self):
        print(f"Account balance for {self.account_owner}: ₦{self.balance}")


def create_account():
    print("\nCreating a new account:")
    account_owner_name = input("Enter your name: ")
    initial_balance = float(input("Enter initial deposit amount: "))
    return Wallet(account_owner_name, initial_balance)


def main():
    print("Welcome to Sonic Trust Bank!")

    user_account = None

    while True:
        print("\nChoose an option:")
        print("1. Create An Account")
        print("2. Make a Deposit")
        print("3. Withdrawal")
        print("4. Check Balance")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            user_account = create_account()
            print(f"Hello, {user_account.account_owner}! Your account has been created with an initial balance of ₦{user_account.balance}.")
        elif choice == '2':
            if user_account:
                deposit_amount = float(input("Enter deposit amount: "))
                user_account.deposit(deposit_amount)
            else:
                print("Please create an account first.")
        elif choice == '3':
            if user_account:
                withdraw_amount = float(input("Enter withdrawal amount: "))
                user_account.withdraw(withdraw_amount)
            else:
                print("Please create an account first.")
        elif choice == '4':
            if user_account:
                user_account.check_balance()
            else:
                print("Please create an account first.")
        elif choice == '5':
            print("Thank you for using the Sonic Trust Bank. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")


if __name__ == "__main__":
    main()
```
